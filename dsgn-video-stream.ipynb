```python
import cv2
import numpy as np
import os
from keras.models import load_model
```

```python
model = load_model("model_checkpoint.keras")
# THIS ONE DOESNT WORK / NEEDS TO BE TRAINED
# model = load_model("final-model.keras") 

model.summary()
```

```python
def assign_occupation(predicted_gender, predicted_race):
    gender_weight = 2
    ethnicity_weight = 1

    # Calculate scores with weights
    occupation_scores = {}
    for occupation, gender_dist in gender_distribution.items():
        race_dist = ethnicity_distribution[occupation]
        gender_score = np.dot(predicted_gender, gender_dist) * gender_weight
        race_score = np.dot(predicted_race, race_dist) * ethnicity_weight
        total_score = gender_score + race_score  # Weighted sum of scores
        occupation_scores[occupation] = total_score

    # Find the occupation with the highest score
    best_occupation = max(occupation_scores, key=occupation_scores.get)

    return best_occupation, occupation_scores
```

```python
# Load the pre-trained face cascade classifier
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Initialize video capture
cap = cv2.VideoCapture(1)  # Changed to capture from the default camera (index 0)

# Check if video capture is opened successfully
if not cap.isOpened():
    print("Error: Failed to open video capture.")
    exit()

# Get the screen resolution
# screen_width = 1921  # Change this to your screen's width resolution
# screen_height = 1202  # Change this to your screen's height resolution

# Set the display window size
# cv2.namedWindow("Fullscreen", cv2.WINDOW_NORMAL)
# cv2.setWindowProperty("Fullscreen", cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)

    # Check if frame is read successfully
    if not ret:
        print("Error: Failed to read frame from video capture.")
        break
    
    # Resize the frame to match the screen resolution
    # frame = cv2.resize(frame, (screen_width, screen_height))

    # Detect faces in the original color frame
    faces = face_cascade.detectMultiScale(frame, scaleFactor=1.2, minNeighbors=8, minSize=(300, 300))

    # Process each detected face
    if len(faces) > 0:
        for i, (x, y, w, h) in enumerate(faces):
            
            # Adjust the face rectangle to zoom into the central part
            adjust_factor = 0.2  # Percentage to zoom
            new_w = int(w * (1 - adjust_factor))
            new_h = int(h * (1 - adjust_factor))
            new_x = x + int(w * adjust_factor / 2)
            new_y = y + int(h * adjust_factor / 2)

            cv2.rectangle(frame, (new_x, new_y), (new_x + new_w, new_y + new_h), (255, 255, 255), 2)
            face_roi_color = frame[new_y:new_y + new_h, new_x:new_x + new_w]

            # Resize the zoomed face region to match the input shape of the model
            resized_face = cv2.resize(face_roi_color, (198, 198))

            # Normalize the face image
            normalized_face = np.array(resized_face) / 255.0  # Corrected normalization of resized_face

            # Reshape the face image to match the input shape of the model
            input_data = np.expand_dims(normalized_face, axis=0)
            
            # Make predictions on the face image (assuming 'model' and 'dataset_dict' are defined elsewhere)
            predictions = model.predict(input_data)
            
            # Extract predictions
            predicted_ethnicity_distribution = predictions[1]
            predicted_gender_distribution = predictions[2]

            best_occupation, occupation_scores = assign_occupation(predicted_gender_distribution, predicted_ethnicity_distribution)
            
            # Output the results
            print("Best matched occupation:", best_occupation)
            print("Scores per occupation:", occupation_scores)

            cv2.putText(frame, best_occupation, (new_x, new_y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 2)

    # Display the resulting frame
    cv2.imshow("", frame)
    
    # Exit the loop if 'Esc' key is pressed
    if cv2.waitKey(1) == 27:
        break

# Release the video capture and close all windows
cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
